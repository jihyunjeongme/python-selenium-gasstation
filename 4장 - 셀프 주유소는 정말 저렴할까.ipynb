```python
# 4-3 구별 주유 가격에 대한 데이터의 정리

import pandas as pd
from glob import glob
```

```python
glob('./data/지역*.xls')
```

```python
# station 변수에 각 엑셀파일의 경로와 이름을 리스트로 저장합니다.

stations_files = glob('./data/지역*.xls')
stations_files
```

```python
# !pip install xlrd

tmp_raw = []

# read_excel로 각 파일을 반복문을 이용해서 읽은 후 tmp_raw 변수에 append 시킵니다.
# 반복문이 끝나고 나면 concat 명령으로 쉽게 하나로 합칠 수 있습니다.
# 방긍금 25개 엑셀 파일을 하나로 합쳤습니다.
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
# 총 545개의 주유소 정보가 저장 된것을 확인
# 가격정보가 숫자형(int, float)이 아니어서 나중에 처리

station_raw.info()
```

```python
station_raw.head()
```

```python
stations = pd.DataFrame({'Oil_store':station_raw['상호'],
                        '주소': station_raw['주소'],
                        '가격': station_raw['휘발유'],
                        '셀프': station_raw['셀프여부'],
                        '상표': station_raw['상표']})

stations.head()
```

```python
# 주소에서 구 이름만 추출 합니다. 
# 그래서 구별 주유 가격도 조사하겠습니다.
# 주소 컬럼의 주소를 봤을 때 빈칸을 기준으로 분리(split)시키고 두 번째 단어를 선택하면 구 이름이 됨

stations['구'] = [eachAddress.split()[1] for eachAddress in stations['주소']]
stations.head()
```

```python
# 일단 head()만 조사했을 때는 이상 없어 보임.
# 5백여 개나 되는 데이터를 다 보기에는 남감
# 이때 unique() 검사를 수행

stations['구'].unique()
```

```python
# 가격이 기록된 컬럼이 숫자형이 아니였다는 것
# 가격이 기록 되지 않은 경우 가 있는지 확인

stations[stations['가격']=='-']
```

```python
# 가격 정보를 입력하지 않은 주유소는 대상에서 제외

stations = stations[stations['가격'] != '-']
stations.head()
```

```python
# 아직 가격 정보가 숫자형으로 변환되지는 않았음.
# 이제 stations의 가격 컬럼은 float 형으로 변경 되었습니다.

stations['가격'] = [float(value) for value in stations['가격']]
```

```python
# 25개의 엑셀을 합쳤기 때문에 index가 중복 될 수 있음.
# reset+index 명령으로 인덱스를 첨음부터 다시 기록
# 그러면 다시 index라는 컬럼이 하나 더 생성되는데 그 부분을 제거

stations.reset_index(inplace=True)
del stations['index']
```

```python
# pivot_table을 이용해서 구격 가격 정보로 변경하고 가격은 평균 값으로 정리

import numpy as np
gu_data = pd.pivot_table(stations, index=['구'], values=['가격'],
                         aggfunc=np.mean)

gu_data.head()
```

```python
# 서울시 구별 정보에 대해 지도로 표현

geo_path = './data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map = folium.Map(location=[37.5502, 126.982], zoom_start=10.5,
                 title='Stamen Toner')
map.choropleth(geo_data = geo_str,
               data = gu_data,
               columns=[gu_data.index, '가격'],
               fill_color='PuRd', #PuRd, YlGnBu
               key_on='feature.id')
map
```

```python
# 4-6 서울시 주유 가격 상하위 10개 주유소 지도에 표기하기
# 주유 가격 상위 10개 주유소를 oil_price_top10이름으로 저장합니다.

oil_price_top10 = stations.sort_values(by='가격', ascending=False).head(10)
oil_price_top10
```

```python
# 하위 10개에 대해서도 oil_price_bottom10에 저장

oil_price_bottom10 = stations.sort_values(by='가격', ascending=True).head(10)
oil_price_bottom10
```

```python
from tqdm import tqdm_notebook

lat = []
lng = []

for n in tqdm_notebook(oil_price_top10.index):
    try:
        tmp_add = str(oil_price_top10['주소'][n]).split('(')[0]
        
        tmp_map = gmaps.geocode(tmp_add)
        
        tmp_loc =tmp_map[0].get('geometry')
        lat.append(tmp_loc['location']['lat'])
        lng.append(tmp_loc['location']['lng'])
        
    except:
        lat.append(np.nan)
        lng.append(np.nan)
        print("Here is nan!")
        
oil_price_top10['lat'] = lat
oil_price_top10['lng'] = lng
oil_price_top10
```
